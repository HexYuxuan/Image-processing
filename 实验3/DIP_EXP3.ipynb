### (1) FFT and IFFT.	
(a) Load the image rhino.jpg, convert it to grayscale.

(b) Perform FFT. Shift the DC component to the center, and show the phase angles and the magnitudes.

(c) Perform IFFT and show the reconstructed image (Tips: remember to shift the DC component back). 

(d) Display the images in the same figure with sub-figures. Add the corresponding title to the sub-figures.

```python
import cv2
import numpy as np
from PIL import Image
import numpy.fft as fp
import matplotlib.pyplot as plt

# (a) Load the image and convert it to grayscale
image = Image.open('images/rhino.jpg').convert('L')

# Convert the image to a NumPy array
image_array = np.array(image)

# (b) Perform FFT
fft = fp.fftshift(fp.fft2(image_array))

# Calculate the phase angles and magnitudes
phase_angles = np.angle(fft)
magnitudes = np.abs(fft)

# (c) Perform IFFT
ifft_shifted = fp.ifftshift(fft)
reconstructed_image = np.abs(fp.ifft2(ifft_shifted))

# (d) Display the images in sub-figures
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

axs[0].imshow(phase_angles, cmap='gray')
axs[0].set_title('Phase Angles')

axs[1].imshow(magnitudes, cmap='gray')
axs[1].set_title('Magnitudes')

axs[2].imshow(reconstructed_image, cmap='gray')
axs[2].set_title('Reconstructed Image')

plt.tight_layout()
plt.show()
```
